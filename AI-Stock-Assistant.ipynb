```python
import yfinance as yf
import pandas as pd
```

```python
# Download stock data (example: Apple)
df = yf.download('AAPL', start='2020-01-01', end='2024-01-01')
print(df.head())
print(df.shape)
# print(df.columns)
# print(df.info())
#dataframe has a MultiIndex column ('Close', 'AAPL')
```

```python
print(df.isnull().sum())
# df[df.duplicated()]#rows all check
# df[df['Close'].duplicated(keep=False)]#in these duplicates should not be dropped.
```

```python
#feature engineering
df.columns = df.columns.get_level_values(0)
# Moving Average (5-day & 10-day)
df['MA5'] = df['Close'].rolling(window=5).mean()#Take last 5 days closing prices and compute average,short trend
df['MA10'] = df['Close'].rolling(window=10).mean()#Take last 10 days closing prices and compute average,longer trend

# Daily Returns
df['Return'] = df['Close'].pct_change()#daily movement,relative to previous day's closing price (Current - Previous) / Previous day's closing price

# Drop NaN values
df = df.dropna()
```

```python
# Create Target (UP/DOWN)
df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)
# print(df.head())
```
